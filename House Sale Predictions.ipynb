# Feature Selection

```python
df = df.drop('id', axis = 1) # the id column is irrelevant to our prediction modelling
```

Would be good if we have the year and maybe the month too

```python
df['date'] = pd.to_datetime(df['date'])
df['yr_sold'] = df['date'].apply(lambda date:date.year)
df = df.drop('date', axis = 1)
```

I decided to drop the zipcode column. It could have been used for further feature engineering, but I am choosing to ignore it since lat and long can already account for it

```python
df['zipcode'].value_counts()
df = df.drop('zipcode', axis =1)
```

Dropping the houses that are greater than $2,000,000. 

```python
df.drop(df[df['price'] > 2000000].index, inplace = True)
```

I wanted to assess the yr_renovated column to evaluate it's worth as a feature for our model. Some houses do have a higher selling price after renovation so it could be worth as a good feature.

# Scaling and Train Test Split

We need to scale the data so they are equivalent in weight for the model

## Train Test Split

```python
X = df.drop('price',axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape #since there are 18 features, we need 18 neurons
```

```python
X_test.shape
```

## Neural Network Model Initialization

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()

model.add(Dense(18, activation = 'relu'))
model.add(Dense(18, activation = 'relu'))
model.add(Dense(18, activation = 'relu'))
model.add(Dense(18, activation = 'relu'))

model.add(Dense(1))
model.compile(optimizer = 'adam' , loss = 'mse')
```

```python
model.fit(x = X_train , y = y_train, validation_data = (X_test,y_test), batch_size = 128, epochs = 400)
```

```python
loss = pd.DataFrame(model.history.history) # allows us to see the model loss vs validation loss. Helps assess overfitting
```

```python
loss.plot()
```

# Prediction and Metrics of Model

```python
y_pred = model.predict(X_test)
```

## Metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
metrics = pd.DataFrame([mean_absolute_error(y_test,y_pred)], 
             columns = ['Mean Squared Error'])
metrics['Root Mean Squared Error'] = np.sqrt(mean_squared_error(y_test,y_pred))
metrics
```

```python
# Plotting the regression
# Our predictions
plt.figure(figsize = (10,4))
plt.scatter(y_test,y_pred)

m, b = np.polyfit(y_test,y_pred, 1)
plt.plot(y_test, m*y_test + b, 'g', label = 'Deep Learning Prediction')

m_lm, b_lm = np.polyfit(y_test,y_pred_lm, 1)
plt.plot(y_test, m_lm*y_test + b_lm, 'k' , label = 'Linear Regression Prediction')

# Perfect predictions
plt.plot(y_test,y_test,'r' , label = 'Perfect Prediction')
plt.legend()
```

## Metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
metrics = pd.DataFrame([mean_absolute_error(y_test,y_pred), mean_absolute_error(y_test,y_pred_lm)], 
             columns = ['Mean Squared Error'], index = ['Deep Learning Model' , 'Linear Regression Model'])
metrics['Root Mean Squared Error'] = [np.sqrt(mean_squared_error(y_test,y_pred)), np.sqrt(mean_squared_error(y_test,y_pred_lm))]
metrics
```

A higher error is observed in the Linear Regression model compared to the Deep Learning model. However, the linear regression model is more accurate with the perfect prediction. Maybe tweaking the deep learning model would help improve its accuracy as well.
